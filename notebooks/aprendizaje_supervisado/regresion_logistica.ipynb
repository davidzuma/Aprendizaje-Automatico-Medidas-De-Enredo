```python
# Importamos las librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
# Creación de datos
class_0_x = np.random.uniform(low=-1, high=0, size=100)
class_0_y = 2*class_0_x +np.random.uniform(low=0.750, high=3, size=100)
class_1_x = np.random.uniform(low=-1, high=0, size=100)
class_1_y = 2*class_1_x +  np.random.uniform(low=-3, high=-0.75, size=100)
class_0_labels = np.zeros(100)
class_1_labels = np.ones(100)

x = np.concatenate((class_0_x, class_1_x), axis=0) 
y = np.concatenate((class_0_y, class_1_y), axis=0)
Y = np.concatenate((class_0_labels, class_1_labels), axis=0)
X = np.stack((x,y),axis = 1)
```

```python
# Visualización de datos
plt.scatter(class_0_x, class_0_y, label = 'Clase 0')
plt.scatter(class_1_x, class_1_y, label = 'Clase 1')
plt.legend()
plt.savefig('classification.jpg')
```

```python
# Entrenamiento y evaluación
clf = LogisticRegression(random_state=1).fit(X, Y)
w = np.hstack((clf.intercept_[:,None], clf.coef_))[0]
w_0,w_1,w_2 = w[2],w[1], w[0]
w_1,w_2
clf.score(X,Y)
```

```python
# Visualización del límite de decisión
plt.scatter(class_0_x, class_0_y, label = 'Clase 0')
plt.scatter(class_1_x, class_1_y, label = 'Clase 1')
f = lambda x: -(w_2*x)/w_1
x = np.linspace(-1, 0, 50000)
plt.plot(x, f(x), color='red',label = 'Límite de decisión')
plt.legend()
plt.savefig('logistic_regression.jpg')
```
